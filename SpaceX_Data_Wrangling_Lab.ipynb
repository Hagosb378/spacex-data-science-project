### Lab 2: Data Wrangling

```
# Install and import required libraries
!pip install pandas
!pip install numpy

import pandas as pd
import numpy as np
```

```
# Load SpaceX dataset from previous lab
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/cognitiveclass/DA0101EN/coursera/project/dataset_part_1.csv")
df.head(10)
```

```
# Identify and calculate the percentage of missing values in each attribute
missing_percentage = df.isnull().sum() / len(df) * 100
missing_percentage
```

```
# Identify which columns are numerical and categorical
df.dtypes
```

### TASK 1: Calculate the number of launches on each site

```
# Calculate number of launches per launch site
launch_counts = df['LaunchSite'].value_counts()
launch_counts
```

### TASK 2: Calculate the number and occurrence of each orbit

```
# Calculate number and occurrence of each orbit
orbit_counts = df['Orbit'].value_counts()
orbit_counts
```

### TASK 3: Calculate the number and occurrence of mission outcomes

```
# Determine number of mission outcomes
landing_outcomes = df['Outcome'].value_counts()
landing_outcomes

for i, outcome in enumerate(landing_outcomes.keys()):
    print(i, outcome)

# Create set of bad outcomes
bad_outcomes = set(landing_outcomes.keys()[[1,3,5,6,7]])
bad_outcomes
```

### TASK 4: Create a landing outcome label (Class column)

```
landing_class = []

for outcome in df['Outcome']:
    if outcome in bad_outcomes:
        landing_class.append(0)
    else:
        landing_class.append(1)

df['Class'] = landing_class
df[['Outcome', 'Class']].head(8)
```

```
# Determine the success rate
success_rate = df['Class'].mean()
print(f"Success Rate: {success_rate * 100:.2f}%")
```

```
# Export the cleaned dataset
df.to_csv("dataset_part_2.csv", index=False)
print("✅ Dataset exported successfully as dataset_part_2.csv")
```
